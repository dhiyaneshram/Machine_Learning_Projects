```python
sns.set_style('darkgrid')
fig, axes = plt.subplots(ncols=3, nrows=4, figsize=(20,15))
axes = axes.flat
for index, column in enumerate(numcols):
    sns.boxplot(
        data=raw_data,
        x=column,
        ax=axes[index]
    )
    axes[index].set_title(column, fontsize=15)
    axes[index].set_xlabel('')
plt.suptitle('Finding Outliers', fontsize=30)
plt.tight_layout()
```
